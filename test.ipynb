```python
import folium
import geopandas as gpd
import pandas as pd
import chardet

import plotly.express
import shapely.geometry
import json
```

```python
larry_data = pd.read_csv("data/state_by_state.csv", encoding="cp1252")
larry_data['STATE'] = larry_data["STATE"].fillna(method='ffill')
larry_data['display'] = 'Potential Funder: ' + larry_data['POTENTIAL FUNDERS '].astype(str) + ', Email: ' + larry_data['EMAIL'].astype(str) + ', Project Description: ' + larry_data['PROJECT DESCRIPTION'].astype(str) + ', Grantees: ' + larry_data['GRANTEES'].astype(str) + ', Link: ' + larry_data['URL'].astype(str)
larry_data = larry_data.drop(columns=['POTENTIAL FUNDERS ', 'EMAIL', 'PROJECT DESCRIPTION','GRANTEES',
                                      'Unnamed: 9','SAMPLE PROJECTS', 'GRANTEE DESCRIPTION ','URL',
                                      'OTHERCONTACT'
                                     ])
larry_data = pd.DataFrame(larry_data.groupby('STATE')['display'].apply(';'.join)).reset_index()
larry_data.head()
```

```python
us_states = gpd.read_file("data/us_states.json")
us_states.rename(columns={'STATE':'STATE_CODE', 'NAME': 'STATE'}, inplace=True)
us_states = pd.merge(us_states, larry_data, left_on="STATE", right_on="STATE")
us_states.shape
us_states.head()

us_states.to_file("data/cleaned.geojson", driver='GeoJSON')
print(us_states.columns)
```
